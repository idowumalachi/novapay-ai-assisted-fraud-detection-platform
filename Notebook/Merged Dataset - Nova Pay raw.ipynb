```python
import pandas as pd

# Load datasets
df_main = pd.read_csv("nova_pay_transcations.csv")
df_boost = pd.read_csv("nova_pay_fraud_boost.csv")

print("Main dataset shape:", df_main.shape)
print("Boost dataset shape:", df_boost.shape)

# Check for column differences
set(df_main.columns) - set(df_boost.columns), set(df_boost.columns) - set(df_main.columns)
```

```python
# align columns
common_cols = df_main.columns.intersection(df_boost.columns)
df_main = df_main[common_cols]
df_boost = df_boost[common_cols]
```

```python
# Concatenate the datasets
df_merged = pd.concat([df_main, df_boost], ignore_index=True)

print("Merged dataset shape:", df_merged.shape)
df_merged.head()
```

```python
# Saved merged Data
df_merged.to_csv("nova_pay_merged_raw.csv", index=False)
print("Merged dataset saved as nova_pay_merged_raw.csv")
```
